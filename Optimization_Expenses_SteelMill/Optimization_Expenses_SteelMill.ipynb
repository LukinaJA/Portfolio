### Обучим модель линейная регрессия

```python
model_lr = LinearRegression()
```

Запустим pipeline, применим масштабирование признаков

```python
pipe = make_pipeline(StandardScaler(), model_lr)

mae = cross_val_score(pipe,
                          features_train,
                          target_train,
                          scoring='neg_mean_absolute_error',
                          cv=5)
```

```python
mae_lr = mae.mean()*-1
```

```python
print(mae_lr)
```

### Обучим модель случайного леса для регрессия

```python
model_rfr = RandomForestRegressor(random_state=80523)
```

Подберем гиперпараметры

```python
params_rfr={'max_depth': range(1, 21, 5),
              'n_estimators': range(30, 160, 30),
              'max_features':range(1, 10, 1),
              'min_samples_leaf':range(1, 5, 1),
              'min_samples_split': range(2, 6, 2)}

rfr = RandomizedSearchCV(model_rfr, params_rfr, n_iter=50, scoring='neg_mean_absolute_error', n_jobs=-1, cv=5)
rfr.fit(features_train, target_train)
model_rfr_best = rfr.best_estimator_
```

```python
model_rfr_best
```

```python
mae_rfr = (rfr.best_score_)*-1
```

```python
print(mae_rfr)
```

### Обучим модель CatBoostR

```python
model_cbr = CatBoostRegressor(random_state=80523, loss_function='RMSE', verbose = False)
```

Используем функцию потерь RMSE - так как задача регрессии

```python
grid = {'learning_rate': [0.03, 0.1, 0.05, 0.01],
        'depth': range(2, 10, 2),
        'l2_leaf_reg': range(1, 6, 2),
        'iterations': [30, 50, 100]}

cbr = RandomizedSearchCV(model_cbr, grid, n_iter=10, cv=5, scoring='neg_mean_absolute_error')
cbr.fit(features_train, target_train, plot=False)
best = cbr.best_estimator_
best_p = cbr.best_params_
print(best, best_p)
```

```python
best_p
```

Рассчитаем МАЕ

```python
mae_cbr = (cbr.best_score_)*-1
mae_cbr
```

### Обучим модель LGBMRegressor

```python
model_LGBM = LGBMRegressor(objective='regression', random_state=12345)
```

```python
param_grid = {
       'learning_rate': [0.1, 0.01],
      'n_estimators': [50, 100, 150],
      'max_depth': [4, 6],
      'colsample_bytree': [0.7, 0.8, 0.9],
      'subsample': [0.7, 0.8, 0.9],
      'min_child_samples': [1, 5, 10]}

lgbm = RandomizedSearchCV(model_LGBM, param_grid, cv=3, n_jobs=-1, n_iter=10, scoring='neg_mean_absolute_error')
```
